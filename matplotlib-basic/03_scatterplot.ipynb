```python
import matplotlib.pyplot as plt
import seaborn as sns

penguins = sns.load_dataset("penguins")

X = penguins["flipper_length_mm"]
Y = penguins["body_mass_g"]

title = "Flipper length(mm) vs Body mass(g)"

plt.xlim(160, 240)
plt.ylim(2500, 6500)
plt.scatter(X, Y)
plt.title(title)

plt.savefig("penguins")
```
