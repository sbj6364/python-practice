```python
import matplotlib.pyplot as plt

X = [0, 1, 2, 3, 4]
Y = [10, 20, 50, 40, 60]
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(X, Y)
plt.subplot(132)
plt.bar(X, Y)
plt.subplot(133)
plt.plot(X, Y)

plt.savefig("graphs")
```

```python
fig, ax = plt.subplots(1,3,figsize=(9,3))

ax[0].scatter(X,Y)
ax[1].bar(X,Y)
ax[2].plot(X,Y)

plt.savefig("subplots1.png")
```
